```python
# import relevant libraries

import numpy as np
import pandas as pd
import os
```

```python
# define path to image dataset
path = './AFRIFASHION1600/'
```

```python
# get image names
images = os.listdir(path)
```

```python
# format name to get image class
def get_style(name):
    names = name.split('_')
    style = ''
    group = names[:-1]
    for idx, i in enumerate(group):
        style += i 
        try:
            group[idx+1]
            style+=' '
        except:
            pass
    return style
```

```python
# test function
get_style('Agbada_118.png')
```

```python
# get image name and class
data= []

for image_id in images:
    image_path = os.path.join(path, image_id)
    style =  get_style(image_id) 
    data.append({ 
                    'image_id': image_id,
                     'clothing': style
                })
```

```python
# read data into dataframe
df = pd.DataFrame(data)
```

```python
# display all rows in dataframe
df.style
```

```python
# show unique clothing 
df['clothing'].unique()
```

```python
# define clothing for males
male = ['African Shirts', 'Agbada', 'Buba and Trouser']
```

```python
# create gender column 
df['gender']= df['clothing'].apply(lambda x: 'Male' if x in male else 'Female')
```

```python
# show unique clothing for male
df[df['gender']=='Male']['clothing'].unique()
```

```python
# show count of each gender class
df.groupby('gender').count()
```

```python
# save df as csv file to directory
df.to_csv('AFRIFASHION1600.csv', index= False)
```
